```python
import seaborn as sns

# Load the 'penguins' dataset
penguins = sns.load_dataset('penguins')

# Display the first few rows
print(penguins.head())
```

```python
# Boxplot of flipper length by species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguins)
plt.title('Flipper Length by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
# Pairplot of penguin features
sns.pairplot(penguins, hue='species')
plt.title('Pairplot of Penguin Features')
plt.show()
```

```python
# Scatterplot of bill length vs. flipper length
plt.figure(figsize=(10, 6))
sns.regplot(x='bill_length_mm', y='flipper_length_mm', data=penguins)
plt.title('Bill Length vs. Flipper Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of flipper length
plt.figure(figsize=(10, 6))
sns.histplot(penguins['flipper_length_mm'].dropna(), kde=True)
plt.title('Distribution of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.show()
```
