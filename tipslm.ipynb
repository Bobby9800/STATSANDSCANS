```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm

# Step 1: Load the tips dataset
tips = sns.load_dataset('tips')

# Step 2: Perform basic data analysis
print(tips.describe())

# Step 3: Plotting distributions of tips and total bill
sns.histplot(tips['total_bill'], kde=True, color='blue')
plt.title('Distribution of Total Bill')
plt.show()

sns.histplot(tips['tip'], kde=True, color='green')
plt.title('Distribution of Tips')
plt.show()

# Step 4: Perform linear regression
X = tips['total_bill']  # Predictor
y = tips['tip']  # Target

X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(y, X).fit()  # OLS Regression
print(model.summary())  # Display regression results

# Step 5: Visualize regression results
sns.regplot(x='total_bill', y='tip', data=tips, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Regression Line for Tips vs. Total Bill')
plt.show()
```
