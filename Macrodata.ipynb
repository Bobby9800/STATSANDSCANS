```python
import statsmodels.api as sm
import pandas as pd

# Load the macrodata dataset
df = sm.datasets.macrodata.load_pandas().data

# Convert the year and quarter to a datetime format
df['year'] = df['year'].astype(int)
df['quarter'] = df['quarter'].astype(int)
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + (df['quarter'] * 3).astype(str) + '-01')

# Display the first few rows
df.head()
```

```python
# Calculate correlation matrix
correlation_matrix = df[['realgdp', 'realcons', 'realinv', 'cpi']].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform the Augmented Dickey-Fuller test
result = adfuller(df['realgdp'])

# Extract and display the results
print('ADF Statistic:', result[0])
print('p-value:', result[1])
for key, value in result[4].items():
    print('Critical Values:')
    print(f'   {key}, {value}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot GDP over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='realgdp', data=df)
plt.title('Real GDP Over Time')
plt.xlabel('Year')
plt.ylabel('Real GDP')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Economic Indicators')
plt.show()
```

```python
import plotly.express as px

# Interactive plot for GDP and Consumption over time
fig = px.line(df, x='date', y=['realgdp', 'realcons'], labels={'value': 'Economic Indicator'}, title='GDP and Consumption Over Time')
fig.show()
```
