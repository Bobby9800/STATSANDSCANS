```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('WWE_History_1000.csv')

# Check the structure of the data
df.head()
```

```python
# Number of ppv wins per wrestler
reigns_per_wrestler = df['Winner'].value_counts().head(10)

plt.figure(figsize=(10, 6))
reigns_per_wrestler.plot(kind='bar', color='skyblue')
plt.title('Top 10 Wrestlers with Most wins')
plt.xlabel('Wrestler')
plt.ylabel('Number of wins')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot championships
championship_count = df['Event'].value_counts().head(10)

plt.figure(figsize=(10, 6))
championship_count.plot(kind='bar', color='green')
plt.title('Top 10 Most Contested event')
plt.xlabel('event')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
