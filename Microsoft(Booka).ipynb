```python
import yfinance as yf
import pandas as pd

# Fetch Microsoft stock data (MSFT)
msft = yf.download('MSFT', start='2023-01-01', end='2023-12-31')

# Display the first few rows of the dataset
msft.head()
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Line chart for closing prices and moving averages
fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Scatter(x=msft.index, y=msft['Close'], name="Close Price"), secondary_y=False)
fig.add_trace(go.Scatter(x=msft.index, y=msft['MA50'], name="50-Day MA"), secondary_y=False)
fig.add_trace(go.Scatter(x=msft.index, y=msft['MA200'], name="200-Day MA"), secondary_y=False)

fig.update_layout(title="Microsoft Stock Price and Moving Averages")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Price (USD)", secondary_y=False)

fig.show()

# Candlestick chart
fig = go.Figure(data=[go.Candlestick(x=msft.index,
                open=msft['Open'],
                high=msft['High'],
                low=msft['Low'],
                close=msft['Close'])])

fig.update_layout(title="Microsoft Candlestick Chart")
fig.show()
```

```python
# Calculate moving averages
msft['MA50'] = msft['Close'].rolling(window=50).mean()
msft['MA200'] = msft['Close'].rolling(window=200).mean()

# Calculate daily returns
msft['Daily Return'] = msft['Close'].pct_change()

# Display the modified dataset
msft.head()
```

```python
import plotly.express as px

# Interactive line chart for daily returns
fig = px.line(msft, x=msft.index, y='Daily Return', title='Daily Return of Microsoft Stock')
fig.show()
```
